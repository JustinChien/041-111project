```python
import LoadingDataset
import utils
import models
import config

from sklearn.metrics import multilabel_confusion_matrix,classification_report,confusion_matrix
import numpy as np
```

```python
X_train, X_test ,y_train , y_test = LoadingDataset.loading_imgs( \
                        IMG_DIR=config.IMG_DIR, IMG_SIZE=config.IMG_SIZE, data_type=config.data_type, show_detail=config.show_detail)
```
