```python
#define some functions

def join_path(root:str , *params)-> str:
    temp = root
    for p in params:
        temp = os.path.join(temp,p)
    return temp

def plot_multi_label_confusion_metrix(cm,classes,cmap=plt.cm.Blues) -> None:
    for i in range(len(cm)):
        temp_cm = cm[i]
        class_ = classes[i]
        plt.figure(figsize=(6,6))
        plt.imshow(temp_cm,interpolation='nearest',cmap=cmap)
        plt.title(class_)
        plt.colorbar()
        plt.xticks(np.arange(2),)
        plt.yticks(np.arange(2),["TrueNegative , FalsePositive","FalseNegative , TruePositive "])
        thresh = cm[i].max() / 2.
        for i,j in itertools.product(range(len(temp_cm[:][0])),range(len(temp_cm[0][:]))):
                plt.text(j,i,temp_cm[i][j],
                        ha = "center",
                        color = "orange" if temp_cm[i,j]> thresh else "purple")
    
def plot_acc_loss(acc,val_acc,loss,val_loss):
    plt.figure(figsize=(8,8))
    
    plt.subplot(2,1,1)
    plt.plot(acc,label="Trainging Accuracy")
    plt.plot(val_acc,label="Validation Accuracy")
    plt.legend(loc="upper left")
    plt.ylabel("Accuracy")
    plt.ylim([min(plt.ylim()),1])
    plt.title("Training and Validation Accuracy")

    plt.subplot(2,1,2)
    plt.plot(loss,label="Trainging Loss")
    plt.plot(val_loss,label="Validation Loss")
    plt.legend(loc="upper right")
    plt.ylabel("Cross Entropy")
    # plt.ylim([0,1.0]) #too high so let's not use this now
    plt.title("Trainging and Validation Loss")
    plt.xlabel("epoch")
    plt.show()
    
def plot_imgs(imgs,labels):
    plt.figure(figsize=(15,15))
    for i in range(len(imgs)):
        ax = plt.subplot(3,3,i+1)
        plt.imshow(imgs[i])
        plt.title(labels[i])
        plt.axis("off")
        
def convert_cm_to_container(cm):
    cm_container = str()
    for i,j,k in itertools.product(range(len(cm)),range(2),range(2)):
        cm_container += "{}_".format(cm[i,j,k])
    return cm_container

def convert_container_to_cm(cm_container):
    temp = cm_container.split(sep="_")
    size = int((len(temp)-1)/4)
    # print(size)
    c_cm = np.empty((size,2,2))
    count = -1
    for i,j,k in itertools.product(range(size),range(2),range(2)):
        count += 1
        # print(count)
        c_cm[i,j,k] = temp[count]
    return c_cm

def show_report(*nums, show_plot:bool):
    for num in nums:
        history_report = pd.read_csv(report_csv_path)
        current = history_report.iloc[num]
        current_cm = convert_container_to_cm(current["Confusion_Metrix"])
        print("---- ---- ---- index:{} ---- ---- ----\n".format(num))
        for column in columns:
            print("{:>12}:{:>12}\n".format(column,current[column]))
        print("{:>12}:\n\n{}".format("Training_Report",current["Training_Report"]))

        if show_plot == True:
            print("{:>12}:\n".format("Confusion_Metrix"))
            # print(current_cm)
            if len(current_cm)==6:
                plot_multi_label_confusion_metrix(cm=current_cm,classes=['Benign', 'CC', 'Calc', 'MLO', 'Mass', 'Melignant'])
            elif len(current_cm)==4:
                plot_multi_label_confusion_metrix(cm=current_cm,classes=['Benign', 'Calc', 'Mass', 'Melignant'])
```

```python
#Variables and Constants
DATA_TYPE = "ALL" #options: ALL,CC,MLO
DATA_CLASS = ['Benign','Malignant']
model_name = "MobileNet" #Options: MobileNet,Inception,VGG,DenseNet
learning_rate = 1e-05
epochs = 20
BATCH_SIZE = 32
IMG_SIZE = (224,224,3)
show_detail = True


X_dataset = list()
IMG_DIR = "D:\\GitHub\\041-111project\\data\\Cropped_Calc_imgs"
report_csv_path = "D:\\GitHub\\041-111project\\data\\history_report_CalcOnly.csv"
data_csv_path = {"ALL":"D:\\GitHub\\041-111project\\data\\Calc_ALL.csv",
                 "CC":"D:\\GitHub\\041-111project\\data\\Calc_CC.csv",
                 "MLO":"D:\\GitHub\\041-111project\\data\\Calc_MLO.csv"}
```

```python
#Prepare data for feeding into the model 


df = pd.read_csv(data_csv_path[DATA_TYPE]) #['Benign', 'CC', 'Calc', 'MLO', 'Mass', 'Melignant']
# DATA_CLASS = list(df.columns)[2:]

for i in tqdm(range(df.shape[0])):
    img = image.load_img(join_path(IMG_DIR ,df['img_name'][i]),target_size=IMG_SIZE)
    img = image.img_to_array(img)
    # img = img/255. #imma use keras layers to preprocess imgs
    X_dataset.append(img)
    
X = np.array(X_dataset)
y = np.array(df.drop(['img_name','Calc/Mass','R/L', 'CC/MLO', 'Benign/Malignant'],axis=1))


# split the data into 70% as training set and 30% as testing set 
X_train, X_test ,y_train , y_test = train_test_split(X,y,random_state=12,test_size=0.3,shuffle = True)
```

```python
# test if data import correctly
if show_detail:
    imgs = X_train[0:9]
    labels = y_train[0:9]
    imgs = imgs/255.
    plot_imgs(imgs,labels)

    imgs_1 = X_test[0:9]
    labels_1 = y_test[0:9]
    imgs_1 = imgs_1/255.
    plot_imgs(imgs_1,labels_1)
```

```python
def densenet():
  data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.5)
    # ,tf.keras.layers.RandomTranslation(
    #     height_factor=(-0.2, 0.3),
    #     width_factor=(-0.2, 0.3),
    #     fill_mode='nearest')
    ],name="Data_augmentation")

  base_model = tf.keras.applications.DenseNet121(input_shape=(IMG_SIZE[0],IMG_SIZE[1],3),include_top=False,weights=None)
  base_model.load_weights("/content/DenseNet-BC-121-32-no-top.h5")
  base_model.trainable = False

  preprocess_input = tf.keras.applications.densenet.preprocess_input

  flatten_layer = tf.keras.layers.Flatten()

  prediction_layer = tf.keras.layers.Dense(len(DATA_CLASS),activation="softmax")

  inputs = tf.keras.Input(shape=(IMG_SIZE[0],IMG_SIZE[1],3))
  x = data_augmentation(inputs)
  x = preprocess_input(x)
  x = base_model(x,training=False)
  x = flatten_layer(x)
  # x = GlobalAveragePooling2D()(x)
  H1 = Dense(256, activation='relu')(x)
  x = BatchNormalization()(x)
  x = Dropout(0.3)(x)
  # x = BatchNormalization()(x)
  H2 = Dense(64, activation='relu')(x)
  x = Dropout(0.3)(H2)
  x = BatchNormalization()(x)
  outputs = prediction_layer(x)
  model = tf.keras.Model(inputs,outputs)

  return model

def densenetori():
  data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2)
    ],name="Data_augmentation")
  preprocess_input = tf.keras.applications.densenet.preprocess_input
  
  preTrainedModelDenseNet169 = tf.keras.applications.DenseNet169(input_shape =((IMG_SIZE[0],IMG_SIZE[1],3)),include_top=False,weights=None)
  preTrainedModelDenseNet169.load_weights("/content/DenseNet-BC-169-32-no-top.h5")
  for layer in preTrainedModelDenseNet169.layers:
    layer.trainable = False  


  #DenseNet169 Model
  inputs = tf.keras.Input(shape=(IMG_SIZE[0],IMG_SIZE[1],3))
  x = data_augmentation(inputs)
  x = preprocess_input(x)
  x = preTrainedModelDenseNet169(x,training=False)
  x = Flatten()(x)
  # x=Flatten()(preTrainedModelDenseNet169.input)

  # x=GlobalAveragePooling2D()(preTrainedModelDenseNet169.output)
  
  #Fully Connection Layers
  # FC1
  x=Dense(256, activation="relu")(x)
  x=BatchNormalization()(x)
  x=Dropout(0.5)(x)
  # x=BatchNormalization()(x)
  # x=Dense(64, activation="relu")(x)
  # #Dropout to avoid overfitting effect
  # x=Dropout(0.2)(x)
  # # # FC2
  x=Dense(64, activation="relu")(x)
  x=BatchNormalization()(x)
  x=Dropout(0.5)(x)
  # x=Dense(10, activation="relu")(x)


  #output layer
  x=Dense(2,activation="softmax")(x)


  model=Model(inputs,x)

  return model

def densenet161():
  data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.5)
    # ,tf.keras.layers.RandomTranslation(
    #     height_factor=(-0.2, 0.3),
    #     width_factor=(-0.2, 0.3),
    #     fill_mode='nearest')
    ],name="Data_augmentation")

  base_model = tf.keras.applications.DenseNet201(input_shape=(IMG_SIZE[0],IMG_SIZE[1],3),include_top=False,weights="imagenet")
  # base_model.load_weights("/content/DenseNet-BC-161-48-no-top.h5")
  base_model.trainable = False

  preprocess_input = tf.keras.applications.densenet.preprocess_input

  flatten_layer = tf.keras.layers.Flatten()

  prediction_layer = tf.keras.layers.Dense(len(DATA_CLASS),activation="softmax")

  inputs = tf.keras.Input(shape=(IMG_SIZE[0],IMG_SIZE[1],3))
  x = data_augmentation(inputs)
  x = preprocess_input(x)
  x = base_model(x,training=False)
  x = flatten_layer(x)
  # x = GlobalAveragePooling2D()(x)
  # H1 = Dense(256,kernel_initializer='he_uniform')(x)
  # x = Activation("relu")(x)
  H1 = Dense(256,activation='relu')(x)
  x = BatchNormalization()(x)
  x = Dropout(0.3)(x)
  # x = BatchNormalization()(x)
  H2 = Dense(32, activation='relu')(x)
  x = BatchNormalization()(x)
  x = Dropout(0.3)(H2)
  # x = BatchNormalization()(x)
  outputs = prediction_layer(x)
  model = tf.keras.Model(inputs,outputs)

  return model

def nasnet():
  #Rescale Pixel Values
  # Data Augmentation
  data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2)
    ],name="Data_augmentation")

  preprocess_input = tf.keras.applications.nasnet.preprocess_input

  base_model = tf.keras.applications.NASNetMobile(input_shape=(IMG_SIZE[0],IMG_SIZE[1],3),include_top=False,weights="imagenet")

  base_model.trainable = False

  # global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
  flatten_layer = tf.keras.layers.Flatten()

  prediction_layer = tf.keras.layers.Dense(len(DATA_CLASS),activation="softmax")

  inputs = tf.keras.Input(shape=(IMG_SIZE[0],IMG_SIZE[1],3))
  x = data_augmentation(inputs)
  x = preprocess_input(x)
  x = base_model(x,training=False)
  x = flatten_layer(x)
  outputs = prediction_layer(x)
  model = tf.keras.Model(inputs,outputs)
  
  return model


def desnet121try():
  from keras.layers import Input, Dense, Flatten, Dropout, LeakyReLU, Activation
  data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.5)
    # ,tf.keras.layers.RandomTranslation(
    #     height_factor=(-0.2, 0.3),
    #     width_factor=(-0.2, 0.3),
    #     fill_mode='nearest')
    ],name="Data_augmentation")

  base_model = tf.keras.applications.DenseNet121(input_shape=(IMG_SIZE[0],IMG_SIZE[1],3),include_top=False,weights=None)
  base_model.load_weights("/content/DenseNet-BC-121-32-no-top.h5")
  base_model.trainable = False

  preprocess_input = tf.keras.applications.densenet.preprocess_input

  flatten_layer = tf.keras.layers.Flatten()

  prediction_layer = tf.keras.layers.Dense(len(DATA_CLASS),activation="softmax")

  inputs = tf.keras.Input(shape=(IMG_SIZE[0],IMG_SIZE[1],3))
  x = data_augmentation(inputs)
  # x = preprocess_input(x)

  # DenseNet
  x = base_model(x)
  # Our FC layer
  flat1 = Flatten()(x)

  dense1 = Dense(units=256, use_bias=True)(flat1)
  batchnorm1 = BatchNormalization()(dense1)
  act1 = Activation(activation='relu')(batchnorm1)
  drop1 = Dropout(rate=0.5)(act1)

  dense1 = Dense(units=32, use_bias=True)(flat1)
  batchnorm1 = BatchNormalization()(dense1)
  act1 = Activation(activation='relu')(batchnorm1)
  drop1 = Dropout(rate=0.3)(act1)
  # Output
  out = Dense(units=2, activation='softmax')(drop1)

  # Create Model
  model = Model(inputs=inputs, outputs=out)

  return model

def choose_model():
  name = input("請輸入你想要執行的模型：1.Nasnet 2.Densenet121 3.Densenet169 4.Densenet161: 5.Desnet121try: ")
  mod = ["Nasnet","Densenet121","Densenet169","Densenet161","Desnet121try"]
  fun = [nasnet(),densenet(),densenetori(),densenet161(),desnet121try()]
  for i in range(len(mod)):
    if name == str(mod[i]):
      print("yes")
      model = fun[i]
  return model,name
```

```python
#Combined models into one cell and use model_name to choose

if model_name == "DenseNet":
    model,name = choose_model()

    if show_detail == True:
        print("{} Model Created.".format(name))
        model.summary()
    
else:     
    # Data Augmentation
    data_augmentation = tf.keras.Sequential([
        tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    ],name="Data_augmentation")

    #--- --- --- --- 分隔線 --- --- --- ---#        

    if model_name == "MobileNet":

        #Rescale Pixel Values
        preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input

        #MobileNet itself
        base_model = tf.keras.applications.MobileNetV2(input_shape = (IMG_SIZE[0],IMG_SIZE[1],3),
                                                        include_top=False,
                                                          weights="imagenet")
        base_model.trainable = False

    #--- --- --- --- 分隔線 --- --- --- ---#        

    elif model_name == "Inception":

        #Rescale Pixel Values
        preprocess_input = tf.keras.applications.inception_v3.preprocess_input
        #MobileNet itself

        base_model=tf.keras.applications.inception_v3.InceptionV3(
                    include_top=False,
                    weights='imagenet',
                    input_tensor=None,
                    input_shape=(IMG_SIZE[0],IMG_SIZE[1],3),
                    pooling=None,
                    classes=1000,
                    classifier_activation='softmax'
        )

        base_model.trainable = False

    #--- --- --- --- 分隔線 --- --- --- ---#        

    elif model_name == "VGG":

        #Rescale Pixel Values
        preprocess_input = tf.keras.applications.vgg19.preprocess_input

        #VGG19 itself
        base_model = VGG19(weights='imagenet',include_top=False)
        base_model.trainable = False   

    #--- --- --- --- 分隔線 --- --- --- ---#        

    # global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
    flatten_layer = tf.keras.layers.Flatten()

    prediction_layer = tf.keras.layers.Dense(len(DATA_CLASS),activation="softmax")


    # # Assemble all together
    inputs = tf.keras.Input(shape=(IMG_SIZE[0],IMG_SIZE[1],3))
    x = data_augmentation(inputs)
    x = preprocess_input(x)
    x = base_model(x,training=False)
    x = flatten_layer(x)
    x = Dense(256,activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    outputs = prediction_layer(x)
    model = tf.keras.Model(inputs,outputs)

    if show_detail == True:
        print("{} Model Created.".format(model_name))
        model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate= learning_rate),loss = tf.keras.losses.binary_crossentropy, metrics=["accuracy"])
```

```python
loss0,accuracy0 = model.evaluate(x=X_test,y=y_test,batch_size=BATCH_SIZE)

print("initial loss: {:.2f}".format(loss0))
print("initial accuracy: {:.2f}".format(accuracy0))
```

```python
history = model.fit(x=X_train, y=y_train, epochs=epochs, validation_data=(X_test,y_test), batch_size=BATCH_SIZE, shuffle=True)
```

```python
predictions = model.predict(x=X_test)
```

```python
y_pred=np.argmax(predictions,axis=-1)
cm = confusion_matrix(y_true=y_test,y_pred=y_pred)
cm_container = convert_cm_to_container(cm)
report = classification_report(y_true=y_test,y_pred=y_pred,target_names=DATA_CLASS)
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
```
